## This notebook contains different approaches to predict class of comment below YT video (is it spam or not). 
## Rerun it again :)

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
all_artist_data = pd.read_csv("all_artists_data.csv")
```

```python
all_artist_data.head(10)
```

```python
mapping = {"SPAM": 1, "VALID": 0}
```

```python
all_artist_data["class_id"] = all_artist_data["CLASS"].apply(lambda x: mapping[x])
```

```python
def check_link_inside_comment(text: str) -> bool:
    if "http" in text:
        return True
    else:
        return False
```

```python
all_artist_data["http"] = all_artist_data["CONTENT"].apply(lambda x: int(check_link_inside_comment(x)))
```

```python
all_artist_data
```

```python
all_artist_data["class_id_equals_http"] = all_artist_data["class_id"] == all_artist_data["http"]
```

```python
all_artist_data
```

```python
all_artist_data["class_id_equals_http"].sum()
```

## Count Vectorizer to transform texts into matrix, counting number of occurences of each word from dictionary

```python
vectorizer = CountVectorizer(min_df=5, max_df=0.75)
```

```python
features = vectorizer.fit_transform(all_artist_data["CONTENT"]).toarray()
```

```python
x = features
```

```python
y = all_artist_data["class_id"]
```

## Train test split to train on one part of sample and verify accuracy on another

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

## Logistic Regression model - to be able to predict class of given text

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
model.predict_proba(x_test)
```

```python
y_test
```

```python
model.score(x_test, y_test)
```

```python
spam_text = "Hey guys check out my youtube channel!!!!"
```

```python
new_feature = vectorizer.transform([spam_text])
```

```python
new_feature.toarray()
```

```python
model.predict_proba(new_feature.toarray())
```
